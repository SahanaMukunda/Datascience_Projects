```python
#libraries
import pandas as pd
import numpy as np
```

```python
#dataset
book=pd.read_csv("/content/book.csv", encoding='latin-1')
book
```

```python
# rename the columns
book.rename({'User.ID': 'userID','Book.Title':'BookTitle','Book.Rating': 'BookRating'},axis=1,inplace=True)
book
```

```python
len(book.userID.unique())
```

```python
len(book.BookTitle.unique())
```

```python
book_df=pd.pivot_table(data=book,index='userID', columns='BookTitle', values='BookRating')
```

```python
book_df.head()
```

```python
book_df.index=book.userID.unique()
book_df
```

```python
#impute the NaNs with 0 values
book_df.fillna(0,inplace=True)
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#store the results in a dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
#set the index and columns names to userID
user_sim_df.index=book.userID.unique()
user_sim_df.columns=book.userID.unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
#get rid of the diagonal values
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
#most similar users
user_sim_df.idxmax(axis=1)[0:5]
```
